```python
def is_royal_flush(hand):
    for i in range(4):
        h = {13*i+8,13*i+9,13*i+10,13*i+11,13*i+12}
        if h.issubset(hand): return h
    return None
```

```python
def is_full_house(hand):
    vs = Counter([h%13 for h in hand]).most_common(2)
    return vs if vs[0][1]>=3 and vs[1][1]>=2 else None
    
```

```python
is_full_house({10,12,36,1,49,14,23})
```

```python
def is_flush(hand):
    hand = sorted(hand, reverse=True)
    vs = Counter([h//13 for h in hand]).most_common(1)
    if vs[0][1]>=5:
        return sorted([h%13 for h in hand if h//13==vs[0][0]], reverse=True)
    else:
        return None
```

```python
is_flush({10,11,3,1,49,12,23})
```

```python
STRAIGHT_OPTIONS = [set(range(r-5,r)) for r in range(5,13+1)] + [{13,0,1,2,3}]
```

```python
def is_straight(hand):
    hand = {h%13 for h in hand}
#     print(hand)
    for s in STRAIGHT_OPTIONS:
        if s.issubset(hand):
            return s
    return None
```

```python
is_straight({10,11,3,1,4,2,6})
```

```python
def is_two_pair(hand):
    vs = Counter([h%13 for h in hand]).most_common(2)
    return vs if vs[0][1]>=2 and vs[1][1]>=2 else None
```

```python
def high(hand):
    hand = [h%13 for h in hand]
    return max(hand)
```

```python
def is_straight_flush(hand):
    hand = is_flush(hand)
    if hand is not None and is_straight(hand):
        return normalize_hand(hand)
```

```python
def check(hands,flop,f):
    results = [(f(hand+flop),i) for i,hand in enumerate(hands)]
    results = [r for r in results if r[0]]
    return sorted(results, reverse=True)[0] if results else [None, None]
```

```python
def get_winner(hands, flop):
#     print('is_royal_flush')
    result = check(hands, flop, is_royal_flush)    
    if result[0] is not None: wins.append('royal flush'); return result[1]
#     print('is_straight_flush')
    result = check(hands, flop, is_straight_flush) 
    if result[0] is not None: wins.append(f'straight flush'); return result[1]
#     print('is_four')
    result = check(hands, flop, is_4)    
    if result[0] is not None: wins.append('4'); return result[1]
#     print('is_full_house')
    result = check(hands, flop, is_full_house)    
    if result[0] is not None: wins.append('full house'); return result[1]
#     print('is_flush')
    result = check(hands, flop, is_flush)    
    if result[0] is not None: wins.append('flush'); return result[1]
#     print('is_straight')
    result = check(hands, flop, is_straight)    
    if result[0] is not None: wins.append('straight'); return result[1]
#     print('is_3')
    result = check(hands, flop, is_3)    
    if result[0] is not None: wins.append('3'); return result[1]
#     print('is_two_pair')
    result = check(hands, flop, is_two_pair)    
    if result[0] is not None: wins.append('2 2'); return result[1]
#     print('is_2')
    result = check(hands, flop, is_2)    
    if result[0] is not None: wins.append('2'); return result[1]
#     print('high')
    result = check(hands, flop, high)    
    wins.append('high'); 
    return result[1]
    
```

```python
def normalize_hand(hand, shape=None):
    return sorted([h%13 for h in hand], reverse=True)
```

```python
def run_iter(num_hands=5):
    s = random.sample(range(13*4), 2*num_hands+5)
#     print(s)
    hands = [[s[i],s[i+1]] for i in range(0,2*num_hands,2)]
#     print(hands)
#     print([[hand[0]%13,hand[1]%13] for hand in hands])
    flop = s[-5:]
#     print(flop)
#     print([f%13 for f in flop])
    winner = get_winner(hands, flop)
#     print(winner)
    for i in range(num_hands):
        hand = tuple(normalize_hand(hands[i])+[hands[i][0]//13==hands[i][1]//13])
        if hand not in d:
            d[hand] = [0,0]
        d[hand][1]+=1
        if i==winner:
            d[hand][0]+=1

```

```python
def gen_df(d, skip_flush=True):
    table = [[None]*13 for i in range(13)]
    for k,v in d.items():
        if k[-1]==skip_flush:
            continue
        table[k[0]][k[1]] = 100*v[0]//v[1]
    cards = list(range(2,11))+['J','Q','K','A']
    df = pd.DataFrame(table, index=cards, columns=cards)
    df = df.replace({None:0})
    df = df.astype(int)
    return df
```

```python
df0 = gen_df(d)
```

```python
def background_gradient(s, m, M, cmap='PuBu', low=0, high=0):
    rng = M - m
    norm = colors.Normalize(m - (rng * low),
                            M + (rng * high))
    normed = norm(s.values)
    c = [colors.rgb2hex(x) for x in plt.cm.get_cmap(cmap)(normed)]
    return ['background-color: %s' % color for color in c]
```
