```python
dfs = pd.read_html('https://www.101languages.net/arabic/most-common-arabic-words/')
```

```python
df = dfs[0]
```

```python
df.head()
```

```python
df['word'] = df.apply(lambda x: x[1].split()[0], axis=1)
```

```python
alphabet_dfs = pd.read_html('https://en.wikipedia.org/wiki/History_of_the_alphabet')
```

```python
alphabet_df = alphabet_dfs[3]
```

```python
alphabet_df.head()
```

```python
alphabet_df[['Hebrew','Arabic']]
```

```python
d = {}
for i in alphabet_df.index[1:]:
    heb = alphabet_df.loc[i, 'Hebrew']
    heb_letters = heb.split()[:-1]
    ar = alphabet_df.loc[i, 'Arabic']
    if type(ar)==float:
        continue
    ar_letters = ar.split(',')
    ar_letters = [a.split()[0].strip('\u200e') for a in ar_letters]
    for a in ar_letters:
        d[a] = heb_letters[0].strip('\u200e')
```

```python
d['Ø§'] = '×'
```

```python
d
```

```python
df.head()
```

```python
word = df.loc[2,'word']
```

```python
trans = ''.join([d[c] if c in d else c for c in word])
```

```python
trans
```

```python
df['trans'] = df['word'].apply(lambda x: ''.join([d[c] if c in d else c for c in x]))
```

```python
df[['in English','trans']].T
```

```python
df[['word', 'in English', 'trans']].to_csv('data/ar_top_100_trans.txt')
```
