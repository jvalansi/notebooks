```python
import os
from collections import Counter
import re
```

```python
dirname = '/Users/jvalansi/Projects/leetcode/algorithms/cpp/'
```

```python
def get_paths(dirname, suffix):
    fpaths = []
    for fpath, subdirs, subfiles in os.walk(dirname):
        for fname in subfiles:
            if fname.endswith(suffix):
                fpaths.append(os.path.join(fpath,fname))
    return fpaths
```

```python
COMMENTS='//.+'
MULTILINE='/\*.+?\*/'
```

```python
def remove_comment_sections(data):
    if MULTILINE is not None:
        c = re.compile(MULTILINE, re.DOTALL)
        data = re.sub(c, '', data)
    if COMMENTS is not None:
        data = re.sub(COMMENTS, '', data)
    return data
```
