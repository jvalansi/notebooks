```python
import networkx
import matplotlib.pyplot as plt
```

```python
def step(prev, graph, all_prev):
    curr = []
    for i in prev:
        if i*2 not in all_prev:
            curr.append(i*2)
            graph.add_edge(i,i*2)
        if (i-1)%3==0 and (i-1)//3 not in all_prev:
            curr.append((i-1)//3)
            graph.add_edge(i,(i-1)//3)
    return curr, graph
```

```python
def print_prev(prev, width=100):
    p = [' ']*width
    for j in prev:
        if j >= width:
            continue
        p[j] = 'X'
    print(''.join(p))
```

```python
def print_graph(graph):
    networkx.draw(graph)
    networkx.draw_random(graph)
    networkx.draw_circular(graph)
    networkx.draw_spectral(graph)
    plt.show()
```

```python
width = 100
prev = [1]
all_prev = prev
graph = networkx.Graph()
for i in range(1000):
    prev, graph = step(prev, graph, all_prev)
    all_prev.extend(prev)
    print_prev(prev)
    networkx.draw(graph)
#     networkx.draw_random(graph)
#     networkx.draw_circular(graph)
#     networkx.draw_spectral(graph)
    plt.show()
```
