```python
import pandas as pd
```

```python
df = pd.read_csv('https://media-cdn.factba.se/rss/csv/calendar.csv')
```

```python
details_df = pd.DataFrame(df['Details'].str.split('[\(\)]', 2, expand=True).values[:,:2], columns=['Details', 'Duration'])
```

```python
df = df.join(details_df, lsuffix='_')
```

```python
df.head()
```

```python
df['Details'].value_counts()
```

```python
from collections import Counter
```

```python
tokens = '\n'.join(df['Details']).split()
```

```python
bigram = [(tokens[i],tokens[i+1]) for i in range(len(tokens)-1)]
```

```python
c2 = Counter(bigram)
```

```python
c2.most_common(100)
```
