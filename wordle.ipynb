```python
def valid(x):
    if type(x)!=str or len(x)!=5:
        return False
    valid_leters = "qidfgjkxvm"
    valid_leters = valid_leters + valid_leters.upper()
    if 'i' not in x or 'd' not in x:
        return False
    
    if not all(c in valid_leters for c in x):
        return False
    return True
    
```
