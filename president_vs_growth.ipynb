# Presidents

```python
import pandas as pd
```

```python
presidents = pd.read_html('https://en.wikipedia.org/wiki/List_of_Presidents_of_the_United_States')
```

```python
presidents[1].dropna(how='all')
```

```python
presidents = pd.read_csv('data/presidents.csv')
```

```python
presidents = presidents.join(presidents['Unnamed: 1'].str.split('\n–\n', n=2, expand=True))
```

```python
presidents[0] = presidents[0].str.split('\n', expand=True)[0]
```

```python
presidents['year'] = pd.to_datetime(presidents[0], format='%B %d, %Y').dt.year
```

```python
presidents = presidents.dropna(subset=['year'])
```

```python
presidents = presidents.set_index('year')
```

```python
presidents = presidents[~presidents.index.duplicated(keep='last')]
```

```python
presidents.index = presidents.index.astype(int)
```

```python
presidents = presidents.reindex(range(presidents.index.min(), 2019), method='ffill')
```

```python
presidents.head()
```

```python
presidents['Unnamed: 6'] = pd.to_numeric(presidents['Unnamed: 6'].replace({'Republican': 1, 'Democratic':-1}), errors='coerse')
```

```python
presidents.join(us_gdp)
```

```python
presidents['gdp'] = us_gdp
```

```python
presidents.tail()
```

```python
presidents['gdp_diff'] = presidents['gdp'].shift(-1)/presidents['gdp']
```

```python
presidents['gdp_diff_diff'] = presidents['gdp_diff'].shift(-1)/presidents['gdp_diff']
```

```python
presidents['Unnamed: 6'].corr(presidents['gdp_diff_diff'])
```
