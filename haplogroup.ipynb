```python
import pandas as pd
import numpy as np
import re
```

```python
y_dna = pd.read_html('https://en.wikipedia.org/wiki/Y-DNA_haplogroups_by_ethnic_group', match='Population')
```

```python
y_dna_df = y_dna[0]
```

```python
y_dna_df.head()
```

```python
def remove_square_brackets(x):
    return re.sub('\[.+\]', '', x) if x==x else x
```

```python
y_dna_df = y_dna_df.applymap(remove_square_brackets)
```

```python
def calc_division(x):
    try:
        x0, x1 = x.split('/')
        if abs(float(x0[-1])/float(x1)*100 - float(x0[:-1]))<1:
            x, x0 = x0[:-1], x0[-1]
        else:
            x, x0 = x0[:-2], x0[-2:]
        return float(x0)/float(x1)*100
    except (ValueError,AttributeError):
        return x
```

```python
y_dna_df = y_dna_df.applymap(calc_division)
```

```python
y_dna_df.head()
```

```python
y_dna_df.mean()
```

```python
cols = y_dna_df.columns[2:]
```

```python
y_dna_df[cols] = y_dna_df[cols].apply(pd.to_numeric, errors='coerce')
```

```python
y_dna_df.head()
```

```python
y_dna_df.mean()
```

```python
y_dna_lang_df = y_dna_df.groupby('Language family[1]').mean()
```
