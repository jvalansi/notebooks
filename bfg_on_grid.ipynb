```python
from collections import deque
```

```python
N = 10
```

```python
a = [[' ']*N]*N
```

```python
class Node:
    def __init__(self, i,j):
        self.i, self.j = i, j
        self.neighbors = set()
    
    def __repr__(self):
        if self.neighbors == set(): return ' '
        if self.neighbors == set([(self.i-1,self.j)]): return '╹'
        if self.neighbors == set([(self.i+1,self.j)]): return '╻'
        if self.neighbors == set([(self.i,self.j-1)]): return '╸'
        if self.neighbors == set([(self.i,self.j+1)]): return '╺'
        if self.neighbors == set([(self.i-1,self.j),(self.i+1,self.j)]): return '┃'
        if self.neighbors == set([(self.i-1,self.j),(self.i,self.j-1)]): return '┛'
        if self.neighbors == set([(self.i-1,self.j),(self.i,self.j+1)]): return '┗'
        if self.neighbors == set([(self.i+1,self.j),(self.i,self.j-1)]): return '┓'
        if self.neighbors == set([(self.i+1,self.j),(self.i,self.j+1)]): return '┏'
        if self.neighbors == set([(self.i,self.j-1),(self.i,self.j+1)]): return '━'
        if self.neighbors == set([(self.i-1,self.j),(self.i+1,self.j),(self.i,self.j-1)]): return '┫'
        if self.neighbors == set([(self.i-1,self.j),(self.i+1,self.j),(self.i,self.j+1)]): return '┣'
        if self.neighbors == set([(self.i-1,self.j),(self.i,self.j-1),(self.i,self.j+1)]): return '┻'
        if self.neighbors == set([(self.i+1,self.j),(self.i,self.j-1),(self.i,self.j+1)]): return '┳'
        if self.neighbors == set([(self.i-1,self.j),(self.i+1,self.j),(self.i,self.j-1),(self.i,self.j+1)]): return '╋'

        
```

```python
for i in range(N):
    for j in range(N):
        a[i][j] = Node(i, j)
```

```python
q = deque([a[N//2][N//2]])
visited = set((N//2,N//2))
while q:
    x = q.popleft()
    print(x.i,x.j)
    print('\n'.join([''.join(str(x)) for x in a]))
    for i,j in [(x.i-1,x.j),(x.i+1,x.j),(x.i,x.j-1),(x.i,x.j+1)]:
        if (i,j) in visited:
            continue
        if not (0<=i<N and 0<=j<N):
            continue
        x.neighbors.add((i,j))
        a[i][j].neighbors.add((x.i,x.j))
        q.append(a[i][j])
        visited.add((i,j))
```
