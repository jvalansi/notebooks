```python
def calc_potential(library, library_books):
    # Calculate the potential profit if signed in today
    N, T, M = list(map(int,library))
    library_scores = [S[int(i)] for i in library_books]
    # sort the books
    library_scores = sorted(library_scores, reverse=True)
    # sum by number of shipped per day
    day_profits = [0]*T
    for d in range(D-T):
        day_profit = sum(library_scores[d*M:(d+1)*M])
        day_profits.append(day_profit+day_profits[-1])
    return pd.Series(reversed(day_profits))
    
    
```
