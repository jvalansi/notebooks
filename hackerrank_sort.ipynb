```python
import re
import pandas as pd
```

```python
def get_raw(fpath='data/hackerrank_scores.txt'):
    with open(fpath) as f:
        data = f.read()

    return re.findall('.+Max Score.+', data)

```

```python
def parse_score(raw_score):
    '''
    >>> score = 'HardMax Score: 60Success Rate: 42.42%'
    >>> parse_score(score)
    ('Hard', 60, 42.42)
    '''
    m = re.search(r'(?P<difficulty>.+)Max Score: (?P<score>\d+)Success Rate: (?P<rate>\d+\.\d+)', raw_score)
    return m.group('difficulty'), int(m.group('score')), float(m.group('rate'))
```

```python
import doctest
doctest.testmod()
```

```python
def sort_results(raw_scores, n=40):
    results = []
    for raw_score in raw_scores:
        results.append(parse_score(raw_score))
    results_df = pd.DataFrame(results, columns=['difficulty', 'score', 'rate'])
    results_df = results_df.sort_values('score', ascending=False)
#     results = [r for r in results if r[0]=='Medium']
    return results_df
```

```python
df = sort_results(get_raw())
```

```python
df['expected'] = df['score']*df['rate']/100
```
