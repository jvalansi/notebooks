```python
import pandas as pd
# df = pd.read_csv('letters.txt', delimiter='\t')
df = pd.read_csv('letters2.txt', delimiter='\t')
df
```

```python
df.rename(columns={k:k.strip() for k in list(df)}, inplace=True)
```

```python
import re
df['Hebrew'] = [re.sub(u'\u202c.+', '', v, flags=re.UNICODE) if not pd.isnull(v) else None for v in df['Hebrew']]
```

```python
def get_dict(df, column1, column2, delim1, delim2, index1=None, index2=0):
    d = {}
    for l, h in zip(df[column1], df[column2]):
        if pd.isnull(l): continue
        letter2 = re.split(delim2, h)[index2].strip() if not pd.isnull(h) else None
        letters1 = re.split(delim1, l)
        if index1:
            d[letters1[index1]] = letter2
        else:
            for letter1 in letters1:
                d[letter1.strip()] = letter2
    return d
# lat2heb = get_dict(df, 'Latin', 'Hebrew', ',', ' ')
lat2heb = get_dict(df, 'Other*', 'Hebrew', ' ', ',', -1)
lat2heb
```

```python
# heb2lat = get_dict(df, 'Hebrew', 'Latin', ',', ' ')
heb2lat = get_dict(df, 'Hebrew', 'Other*', ',', ' ', 0, -1)
heb2lat
```
